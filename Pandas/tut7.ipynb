```python
# Fillna 
import pandas as pd
var2 = pd.read_csv("./tut7a.csv")
var2
var2.fillna("Python")   #Fill the NaN values with "Python"
var2.fillna({'AVG. VOLUME':"python", "TODAY-VOLUME": "c"})  # Fill the NaN values of particular columns with different values
var2.fillna(method= "ffill")  # Fill the NaN values with the value of previous row
var2.fillna(method= "bfill", axis = 1)  # Fill the NaN values with the value of next column
var2.fillna(12, inplace= True) # 12 will be filled inplace of NaN value
var2.fillna('python', limit = 2)  # It replace first 2 NaN values of columns with python
```
