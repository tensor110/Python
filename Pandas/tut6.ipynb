```python
import pandas as pd
csv1 = pd.read_csv("D:\\Python\\Pandas\\tut5c.csv")
csv1
csv1.index   # Returns starting and ending of index
csv1.columns # Returns all the columns head
csv1.describe() # Returns mean, minimum, maximum etc. values
csv1.head()   #Returns 5 rows from top
csv1.head(2)   #Returns 2 rows from top
csv1.tail()   #Returns 5 rows from bottom
csv1.tail(2)   #Returns 2 rows from bottom
csv1[2:5]  # Returns from index 2 to index 4 rows
csv1.index.array  # Returns indexes in array
csv1.to_numpy()  # Returns rows in numpy array format
```
