```python
# Handling misssing values 
# Replace -> replace(to_replaced = val1, value = val2)
import pandas as pd
import pandas as pd
var = pd.read_csv("./tut8a.csv")
var.replace(to_replace=4 , value=116651)  #It replace 1 with 116651
var.replace(to_replace=[64,65,66] , value=60)  #It replace 64,65,66 with 60
var.replace("[A-K]", "pyth", regex= True)  #It replace "A-K" with "pyth"
var.replace({"SYMBOL":"[A-E]"},33, regex= True)  #It replace the value of SYMBOL with 33 which contain any of the character A-E

```
