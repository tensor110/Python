```python
# pie chart 
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3,0.0,0.0,0.0]
# plt.pie(x, labels=y, explode=ex, autopct='%0.1f%%', shadow=True, radius= 1.2)
plt.pie(x, labels=y, labeldistance= 0.6, startangle=90, textprops={'fontsize':15},counterclock=False, wedgeprops= {'linewidth':3, "edgecolor": 'm'}, rotatelabels= True )
plt.legend(loc = 2)
plt.show()
```

```python
# Dot pie chart 
plt.pie([1])
plt.show()

# Ring pie chart 
x = [10,20,30,40]
y = ['c', 'c++', 'java', 'python']
plt.pie(x, labels= y, radius=1.5)
# plt.pie([1], colors='w')
# or 
c = plt.Circle(xy = (0,0), radius= 1, facecolor = 'w')
plt.gca().add_artist(c)
plt.show()
```
