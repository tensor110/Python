```python
# stem plot
import matplotlib.pyplot as plt
x = [1,2,8,6,4,5]
y = [2,5,6,1,2,5]
plt.stem(x,y, linefmt=":", markerfmt='r+', basefmt='g', bottom =2, label = 'python',orientation= 'horizontal')
plt.legend()
plt.show()
```

```python
# box plot 
x = [10,20,30,40,50, 120]
plt.boxplot(x, notch= True, vert = False, widths=0.5, labels = ["python"], patch_artist= True, showmeans= True, whis = 20, sym= 'g+', boxprops= dict(color = 'r'), capprops= dict(color = 'r'), whiskerprops=dict(color = 'r'), flierprops= dict(markeredgecolor = 'y'))
plt.show()
```

```python
# Multiple box plot 
x = [10,20,30,40,50,60]
x1 = [20,50,40,30,50,60]
y = [x , x1]
# plt.boxplot(x)
# plt.boxplot(x1)
plt.boxplot(y, labels= ['python', 'c++'])
plt.show()
```
