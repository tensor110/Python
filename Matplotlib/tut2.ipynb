```python
# Bar Plot
import matplotlib.pyplot as plt
x = ['a', 'b', 'c', 'd']
y = [75,60,80,95]
plt.xlabel("language")
plt.ylabel('No.')
plt.title("Bar Graph", fontsize = 20)
# plt.bar(x,y, width = 0.4, color = "y")
c = ['y', 'b', 'r', 'g']
# plt.bar(x,y, width = 0.4, color = c, align = 'edge', edgecolor = 'b', linewidth = 5, linestyle = ":", alpha = 0.4)
plt.bar(x,y, width = 0.4, label = "popularity")
plt.legend()
plt.show()
```

```python
import numpy as  np

x = ['a', 'b', 'c', 'd']
y = [75,60,80,95]
z = [20,30,40,50]

width = 0.2
p= np.arange(len(x))
p1 = [j+ width for j in p]

plt.xlabel("language")
plt.ylabel('No.')
plt.title("Bar Graph", fontsize = 20)

plt.bar(p,y, width, label = "popularity")
plt.bar(p1,z, width, label = "popularity1")
plt.xticks(p+width/2, x, rotation = 20)

plt.legend()
plt.show()
```

```python
# Horizontal Bar plot 
# import matplotlib.pyplot as plt
x = ['a', 'b', 'c', 'd']
y = [75,60,80,95]
plt.xlabel("language")
plt.ylabel('No.')
plt.title("Bar Graph", fontsize = 20)
plt.barh(x,y ,label = "popularity")
plt.legend()
plt.show()
```
