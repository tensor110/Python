```python
# area or stack plot 
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area = [5,2,1,3,1]
area1 = [1,2,5,3,2]
area2 = [4,3,4,5,3]
l = ['area1', 'area2', 'area3']
plt.stackplot(x,area, area1, area2, labels=l, baseline= 'sym')
plt.legend(loc = 2)
plt.show()
```

```python
# step plot
x = [1,2,3,4,5]
y = [5,2,1,3,1]
plt.step(x,y, color= 'r', marker = 'o', ms = 10, mfc = 'g', label = 'python')
plt.legend(loc = 1)
plt.show()
```

```python
# fill between plot
import numpy as np 
x = np.array([1,2,3,4,5])
y = np.array([5,2,1,3,1])
plt.plot(x,y)
# plt.fill_between(x = [2,4], y1 = 2, y2 =4)
plt.fill_between(x,y, where=(x>=2) & (x<=5))
plt.show()
```
