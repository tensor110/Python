```python
# sub plot -> subplot(rows, columns, index)
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.pie([1])

z = [10,20,30,40]
plt.subplot(2,2,3)
plt.pie(z)

x1 = ['a', 'b', 'c', 'd']
y1 = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x1,y1)
```
