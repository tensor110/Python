```python
# Scatter Plot 
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,2,4,1,3]
sizes = [100,200,300,100,300]
colors = [10,50,90,40,70]
plt.scatter(x,y, s = sizes, c= colors, marker='*', cmap = 'BrBG')
t = plt.colorbar()
t.set_label("ColorBar", fontsize = 20)
plt.show()
```
