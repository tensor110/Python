```python
# Download the plot 
import matplotlib.pyplot as plt
x = [1,5,3,4]
y = [1,2,3,4]
plt.plot(x,y)
plt.savefig('tut8fig.pdf', dpi = 3000, facecolor = 'y', transparent = True, bbox_inches = 'tight')
plt.show()
```

```python
# Changing axis data 
x = [1,5,3,4]
y = [1,2,3,4]
plt.plot(x,y)
# plt.xticks(x, labels=['python', 'c', 'c++', 'java'])
# plt.yticks(x)

plt.xlim(0,10)
plt.ylim(0,10)

plt.axis([0,10,0,7])
plt.show()
```

```python
# Text inside plot 
x = [1,5,3,4]
y = [1,2,3,4]
plt.plot(x,y)

plt.title("Plot", fontsize = 25)
plt.xlabel("Days", fontsize = 10)
plt.ylabel("No.s", fontsize = 10)
# text(pos-x, pos-y, text)
plt.text(1, 3.5, 'text', fontsize = 10, style ='italic', bbox={'facecolor':'red'})
plt.annotate('Ann', xy = (1,1), xytext=(2,2),arrowprops=dict(facecolor = 'black', shrink =100))
plt.legend(['up'], loc = 1, facecolor = 'red', edgecolor = 'yellow', framealpha = 0.8, shadow = True)
plt.show()
```
